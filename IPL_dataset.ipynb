```python
# we are adding new column season to extact year fron date column

match_data['season']=pd.DatetimeIndex(match_data['date']).year
match_data.head(2)
```

```python
#how many matches played in each season 

match_per_season=match_data.groupby(['season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
# ploting a graph for above dataset

sns.countplot(match_data['season'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('season',fontsize=10)
plt.ylabel('count',fontsize=10)
plt.title('Total Matches Played In Each Season',fontsize=10,fontweight='bold')
```

```python
# we merging two table match_data and ball_data

season_data=match_data[['id','season']].merge(ball_data,left_on='id',right_on='id',how='left').drop('id',axis=1)
season_data
```

```python
#WE are displaying total run in each sesson by line graph

season=season_data.groupby(['season'])['total_runs'].sum().reset_index()
p=season.set_index('season')
ax=plt.axes()
ax.set(facecolor='white')
sns.lineplot(data=p,palette='magma')
plt.title('Total Runs In Each Season',fontsize=12,fontweight='bold')
plt.show()
```

```python
#here we are finding out the runs socered per match by dividing the total runs by matches

run_per_season=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
run_per_season['Runs socred per match']=run_per_season['total_runs']/run_per_season['matches']
run_per_season.set_index('season',inplace=True)
run_per_season
```
